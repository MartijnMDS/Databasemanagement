```python
import pandas as pd

path = "C:\\Users\\ginok\\Documents\\Digital Driven Business\\Database management\\metaCleanV1 - kopie.xlsx"

df = pd.read_excel(path)

df2 = pd.DataFrame(df['moviecast'])
```

```python
df2 = df2.dropna()
df2
```

```python
df2['actors'] = df2['moviecast'].str.split(',')
df3 = pd.DataFrame(df2['actors'].explode())
df3
```

```python
df3 = df3.drop_duplicates()
df3
```

```python
df3 = df3.reset_index(drop= True)
```

```python
#add an Id number to each actor using the index +1
df3['Actor_ID'] = df3.index + 1
df3
```

```python
#reorder the columns so the Actor_ID (primary key) is the first column
df3 = df3 [['Actor_ID', 'actors']]
df3
```

```python
df3.to_csv('filename.csv', index=False)
```
