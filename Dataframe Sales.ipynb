```python
# import pandas
import pandas as pd
```

```python
# Import the excel data from Brightspace
df= pd.read_excel(r"C:\Users\Legion 5\Downloads\sales (1).xlsx")
df
```

```python
# Deleting columns that we don't use
df = df.drop(columns=['year', 'release_date', 'title', 'international_box_office', 
                    'domestic_box_office', 'Unnamed: 8', 'theatre_count', 'avg run per theatre', 'keywords', 'creative_type' ])
```

```python
# split the url and change de data type
movie_name= df.url.str.split('/', expand=True)
movie_name
```

```python
#Add the names of the movie to the file with the name "title"
df["title"] = movie_name[:][4]
df
```

```python
# Make url the first column 
first_column = df.pop('url')
df.insert(0, 'url', first_column)
```

```python
# Removing the NaN and change it to 0.0
df= df.fillna(0)
df
```

```python
# Export to CSV file
df.to_csv('', index=False)
```
